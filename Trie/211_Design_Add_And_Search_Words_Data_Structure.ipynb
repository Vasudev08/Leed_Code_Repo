```python
"""
UMPIRE:
U: We are to contruct a Trie and the unquie requirement is that the
word might contain '.' which can be matched with any letter

M: Trie

P: 
NOTE meat of the implementation is handling the '.' where we will use dfs
the implementation is little bit tricky

In the implementation we will have two parts:
1. Recursion (DFS) - when we encounter '.' character
2. Iterative - When we encounter normal character

for 1. we will construct an dfs recursion taking in the index and curNode,
where we will use 'for' loop to traverse through all the childrens to know if
there was indeed an match or not
"""

class TrieNode:
    def __init__(self):
        self.children = {}
        self.isWord = False

class WordDictionary:

    def __init__(self):
        self.root = TrieNode()
        

    def addWord(self, word: str) -> None:
        cur = self.root

        for c in word:
            if c not in cur.children:
                cur.children[c] = TrieNode()

            cur = cur.children[c]

        cur.isWord = True
        

    def search(self, word: str) -> bool:
        def dfs(j, node):
            cur = node

            for i in range(j, len(word)):
                c = word[i]
                if  c == ".":
                    for child in cur.children.values():
                        if (dfs(i + 1, child)):
                            return True
                    return False         
                else:
                    if c not in cur.children:
                        return False

                    cur = cur.children[c]
            return cur.isWord

        return dfs(0, self.root)
        


# Your WordDictionary object will be instantiated and called as such:
# obj = WordDictionary()
# obj.addWord(word)
# param_2 = obj.search(word)
```
