```python
# 767. Reorganize String

# The only move we have is to swap character, 
# No captial letters and stuff like that

# s = "aaabc" => "abaca" Yes valid arragment

# UMPIRE
# Slidling Window or window, with a count of unique characters (only occur once) and duplicated characters(occure more than once)
# for n dupcliated charactesr which should have n - 1 uniquce charaters to reorganize it
# {a:3}, [a, a, a]
# {b:1, c: 1} [b, c ]
# unique = 2
# duplicate= 3
# total = unique + duplication

# curStr = ""
# # for i in range(total):
#     if i % 2 ==0:
#         curStr += unique.pop()
#     else:
#         curStr += duplicate.pop()

# return curStr


class Solution:
    def reorganizeString(self, s: str) -> str:
        occurence = defaultdict()
        uniqueList = []
        duplicateList = []

        uniqueCount = 0
        duplicateCount = 0

        for c in s:
            if occurence[c] > 1:
                uniqueCount -= 1
                duplicateCount += 1

            occurence[c] += 1
            uniqueCount += 1

        if uniqueCount < duplicateCount - 1:
            return ""
        
        curStr = ""
        for i in range(uniqueCount + duplicateCount):
            if i % 2 == 0:
                curStr += uniqueList.pop()
            else:
                curStr += duplicateList.pop()


        return curStr



```

```python
class Solution:
    def reorganizeString(self, s: str) -> str:
        freq = defaultdict(int)

        for ch in s:
            freq[ch] += 1

        maxHeap = []
        heapq.heapify(maxHeap)

        for ch in freq:
            heapq.heappush(maxHeap, (-freq[ch], ch))

        res = []

        while len(maxHeap) >= 2:
            freq1, char1 = heapq.heappop(maxHeap)
            freq2, char2 = heapq.heappop(maxHeap)

            res.append(char1)
            res.append(char2)

            if -freq1 > 1:
                heapq.heappush(maxHeap, (freq1 + 1, char1))

            if -freq2 > 1:
                heapq.heappush(maxHeap, (freq2 + 1, char2))

        
        if maxHeap:
            freq, char = maxHeap[0]
            if -freq > 1:
                return ""
            res.append(char)

        return ''.join(res)
        
# Time: O(n log K)
# Space: O(K)

            

         
```
