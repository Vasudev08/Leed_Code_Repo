```python
"""
I like this problem
2nd Try:
- In Sliding Window approach we moev the window i.e. 
    - increase it or decrease it
    - we don't remake/ restart the window as long

And in this problem, we either decrease or increase the window
using two hashmap and two variables

O(N) running time only

using for and while loop


The above a good overview on how to approach the problem. Now I will into details
1. Create two hashamp countT and window
    - countT count the occurence of each character in t string
    - window will be the count of the occurence of each character in s string
        - if an character is prensent in the countT. we will compare
            the occurence on both hashmap
            - and if occurencce is equal then increase the have
        - Once both need and have are equal we store the mini string 
            - to do this, we are going to decrease the lenght of the window
                until our have decreases. after which we go back into the 
                loop
2. Create two varialbes need and have
    - 
"""

class Solution:
    def minWindow(self, s: str, t: str) -> str:
        if t == "":
            return ""
        
        countT, window = {}, {}
        need, have = len(countT), 0
        res, resLen = [-1, -1], float("infinity")
        l = 0

        # Counting the occurence of each character in t
        for c in t:
            countT[c] = 1 + countT.get(c, 0)

        for r in range(len(s)):
            window[s[r]] = 1 + window.get(s[r], 0)

            if s[r] in countT and window[s[r]] == countT[s[r]]:
                have += 1

            while have == need:
                if r - l + 1 < resLen:
                    res = [l, r]
                    resLen = r - l + 1

                window[s[l]] -= 1
                if s[l] in countT and window[s[l]] < countT[s[l]]:
                    have -= 1
                

                l += 1

        l, r = res
        return s[l: r + 1] if resLen != float("infinity") else ""
```
