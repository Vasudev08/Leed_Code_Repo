```python
class MedianFinder:
    """
    UMPIRE:
    U: Here our primary job is to return the Median
    M: minHeap and maxHeap, where we will have left section represented by maxHeap
        and right section represented by minHeap
        Left Section -> MaxHeap
        Rigth Section -> MinHeap

    P: 
    - by default we will add num to the left section 
    - but we have make sure 
        1. last elem of left section < first element of the right section
            maxHeap(0) -> root node          minHeap(0) -> root node
        2. len of the left section can only be + 1 to the len of the right section

    """

    def __init__(self):
        self.left = []
        self.right = []
        # heapq.heapify(self.left)
        # heapq.heapify(self.right)

    def addNum(self, num: int) -> None:
        heapq.heappush(self.left, -1 * num)

        # Making sure that self.left[0] < self.right[0]
            # otherwise we rearrange it

        if (self.left and self.right and (-1 * self.left[0] > self.right[0])):
            cur = -1 * heapq.heappop(self.left)
            heapq.heappush(self.right, cur)

        if len(self.left) > len(self.right)  + 1:
            cur = -1 * heapq.heappop(self.left)
            heapq.heappush(self.right, cur)

        if len(self.right) > len(self.left) + 1:
            cur = heapq.heappop(self.right)
            heapq.heappush(self.left, -1 * cur)

    def findMedian(self) -> float:
        if len(self.left) > len(self.right):
            return -1 * self.left[0]

        if len(self.right) > len(self.left):
            return self.right[0]

        return (-1 * self.left[0] + self.right[0]) / 2
        


# Your MedianFinder object will be instantiated and called as such:
# obj = MedianFinder()
# obj.addNum(num)
# param_2 = obj.findMedian()
```
