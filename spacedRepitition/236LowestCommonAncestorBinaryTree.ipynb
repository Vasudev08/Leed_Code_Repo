```python
"""
UMPIRE:

U: we are given a binary tree, our job is to return LCA of given two nodes
M: DFS algo

P:
- there are two ways in which their could be a possible LCA
    - case 1: either of nodes 'p' or 'q' is the LCA
    - case 2: Neither node 'p' nor 'q' is the LCA
- we will call dfs function at every node to see if the current node is the LCA

"""

class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        if not root or p == root or q == root:
            return root
        
        left = self.lowestCommonAncestor(root.left, p, q)
        right = self.lowestCommonAncestor(root.right, p, q)

        if left and right:
            return root
        
        return left if left else right
```
