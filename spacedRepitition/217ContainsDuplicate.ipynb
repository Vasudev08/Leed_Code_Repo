### 217. Contains Duplicate - 10/27/2025

```python
# High - Level overview
# we are returing true if an array is duplicated 
# we are returning false if array is unique

# if element no in our set of unique nums, then add to the set
# if the current element is present in unique nums set return True


class Solution(object):
    def containsDuplicate(self, nums ):
        """
        :type nums: List[int]
        :rtype: bool
        """

        uniqueSet = set()
        for i in nums:
            if i not in uniqueSet:
                uniqueSet.add(i)
            else:
                return True
            
        return False
```
