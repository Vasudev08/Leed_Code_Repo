```python
# In this questions we do have k which incidates sliding window

# l = 0
# r = 0

"""
nums = [1, 2, 3, 1], k = 3

I know it's a sliding window question, but I don't know 
what is the best way to traverse this list called nums 
how should I traverse it tho, 
- like keep on incrase r then j - i == k, then we move on
to increase the value l and until l == r, then we move r

I see, I got the questions in mind
basically, for each index we are going to expand k element 
and once we reach k element, then we are going to increment pointer l

pointer -> l will be controlled via the use of while loop
pinter -> r will be controlled via the use of the for loop

"""

class Solution(object):
    def containsNearbyDuplicate(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: bool
        """
        l = 0
        resCount = 0

        while (l < len(nums)):

            for r in range(l + 1, l + k + 1):
                if nums[l] == nums[r]:
                    resCount += 1

            l += 1

        if resCount ==  2:
            return True

        return False
```

```python
class Solution(object):
    def containsNearbyDuplicate(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: bool
        """
        dic = {}

        for i in range(len(nums)):
            if nums[i] in dic and i - dic[nums[i]] <= k:
                return True

            dic[nums[i]] = i
            # Here key: num_value, value = idx

        return False
```
