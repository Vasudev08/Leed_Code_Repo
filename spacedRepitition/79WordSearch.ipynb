```python
"""
UMPIRE

U: we are given mxn grid, where each cell has an alphabet 
we are to return true or false whether or not that word exists
in the grid

M: backTracking, which mean poping the value after calling the function

P:
- we will have two 'for' loops which traverse through the grid,
    and find the first matching letter
- after which we are gonna all the backTracking function. which checks and 
calls the backtracking function twice
    - 1. calling it horizontally
    - 2. popping the last element and then calling it vertically

"""

from typing import List


class Solution:
    def exist(self, board: List[List[str]], word: str) -> bool:
        res = ""
        resBool = False

        def backTracking(i, j, strIndex, res):
            if board[i][j] != word[strIndex] or strIndex >= len(word):
                return
            
            res.append(board[i][j])

            if strIndex == len(word):
                if res == word:
                    resBool = True
                    return 
                
            backTracking(i + 1, j, strIndex + 1, res)
            backTracking(i, j + 1, strIndex + 1, res)



            res.pop()
            backTracking(i, j + 1, strIndex + 1, res) 
            backTracking(i + 1, j, strIndex + 1, res)      


        return resBool      

```
