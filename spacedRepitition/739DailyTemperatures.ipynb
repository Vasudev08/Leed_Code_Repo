```python
def dailyTemperature(temperatures):
    n = len(temperatures)

    reslist = [0] * n
    stack = []

    for i in range(n):
        while stack and temperatures[i] > temperatures[stack[-1]]:
            prev = stack.pop()
            reslist[prev] = i - prev

        stack.append(i)
```
