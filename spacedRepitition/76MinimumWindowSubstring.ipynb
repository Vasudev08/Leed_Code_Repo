```python
"""
UMPIRE:

U:
given two string:
- s
- t

find a minimum length window in 's' which contains all the characters of 't'

this quetions build upon the 424 long repeating character replacement

M: two hashmaps, two varibles 

P:

1. record the occurence of each character in 't', which will also have a variable named 'want'
    this tells no. of characters in the 't' string
2. we will be controlling the sliding window via a 'for' loop, where 
    we keep on increase the length of the window until 'have' variale is equal to 'want' 
    Here have variable will be increased, when we encoutner an charcter in both our window substring and 't'
3. Also once our 'have' and 'want' variable are equal, we will then proceed to incease our left pointer
    until 'have' and 'want' are not equal, then we will increase our right pointer and the process 
    repeats until r reaches the end of the string
4. throughout the above process we are keeping track of minLen and minString and return them 
    at the end.



"""

class Solution:
    def minWindow(self, s: str, t: str) -> str:
        # sliding window technique
        # l = 0, r controlled by for loop
        # Two hashmap + Two Variables
       #Q1. if we are using two hashmaps, how do we know 
       # when the 's' hashmap has all the 't' character
        #A1. Using variables


        """
        increase the variable when the character is in t
        when 'is' and 'want' variable count is same
        store and compare the min lengtht

        """

        tCount = {}
        tWant = len(t)

        winCount = {}
        winHave = 0

        minLen = float("infinity")
        minString = ""

        for c in t:
            tCount[c] = 1 + tCount.get(c, 0)

        l = 0

        for r in range(len(s)):
            if s[r] in tCount:
                winCount[s[r]] = 1 + winCount.get(s[r], 0)
                winHave += 1

            
            while (winHave == tWant):
                if s[l] in tCount:
                    winCount[s[l]] = winCount.get(s[l], 0) - 1
                    

                    winHave -= 1

                if (r - l + 1) < minLen:
                    minLen = r - l + 1
                    minString = s[l: r + 1]

                # minLen = min(minLen, r - l + 1)
                l += 1

        return minString

# NOTE In the above implementation we are not checking or taking into account the count of the character, 
    # sometimes the required count of the character is 2 or 3, but in our code we are only checking if it exists or not
    # we are doing similar while decreasing the winHave variable, only checking the existence of the character, not the count 
    # whether it's less than the count in the tCount hashmap/dictionary
```

```python
class Solution:
    def minWindow(self, s: str, t: str) -> str:
        resString = ""
        resLen = float("infinity")

        tCount = {}
        need = len(t)

        sCount = {}
        have = 0

        l = 0
        for i in range(len(t)):
            tCount[t[i]] = 1 + tCount.get(t[i], 0)

        
        for r in range(len(s)):
            if s[r] in tCount:
                sCount[s[r]] = 1 + sCount.get(s[r], 0)

                if sCount[s[r]] <= tCount[s[r]]:
                    have += 1

            while (have == need):
                if s[l] in tCount:
                    sCount[s[l]] -= 1

                    if sCount[s[l]] < tCount[s[l]]:
                        have -= 1

                
                if (r - l + 1) < resLen:
                    resLen = r - l + 1
                    resString = s[l: r + 1]

                l += 1

        return resString
        
```
