```python
"""
UMPIRE

U: we are given a string 's', return all possible palindrome partioining of 's'

M: in this question we will explore all the combinations so we have 
to use backTracking

P:
Q1: what way should I partition the string 's' and how
possible solution: start to check if a window of len = 1
is a palindrome while traversing through the stirng 's'
then keep on increasing the size of the window
this way we have to create a helper function
which check if the curString a palindrome or not

create a 'for' loop which traverses
- from len == 1 to the len == len(s) -1 which would the window size
- using the given window size we check if the given window is a palindrome
    or not. if so, append to the tempRes list 
    if, current windows size  > renaming length of the string
        - just check if the remaining length is a palnindrome or not
- to check if the currentString a palindrome or not, we create a helper function
- which return true if the currentWindow string a palindrome or not







"""

class Solution:
    def partition(self, s: str) -> List[List[str]]:

        res = []
        def checkPalin(curStr):
            l = 0
            r = len(curStr) - 1


            while (l < r):
                if s[l] != s[r]:
                    return False
                
                l += 1
                r -= 1


            return True
        

        for winSize in range(1, len(s) + 1):
            startIndex = 0
            tempRes = []
            while (winSize + startIndex) < len(s):
                if checkPalin(s[startIndex: startIndex + winSize]):
                    tempRes.append(s[startIndex: startIndex + winSize])

                startIndex += 1

            if checkPalin(s[startIndex:]):
                tempRes.append(s[startIndex:])

            res.append(tempRes)

        return res


# NOTE: close yet to far, I will complete this question tomorrow
# this is a pretty pretty good question for backtracking
# I would love to do it again

```

```python
"""2nd Repetition after a lot of help

UMPIRE

U: we are given a string, we are to return the all palindrome partition of the given string 's'

M: Backtracking to get access all the parition 

P: 
1. we will create two helper function, we will also have two list
    - def checkPali(curStr)
    - def backTracking(index)
    - res 
    - curParition

2. Here, the checkPali function will check if a given string a palindrome or not
3. for backTraking fucntion implementation please check the code implementation below


"""

class Solution:
    def partition(self, s: str) -> List[List[str]]:
        res = []
        curPart = []


        def checkPali(curStr):
            l = 0
            r = len(curStr) - 1

            while (l < r):
                if curStr[l] != curStr[r]:
                    return False
                
                l += 1
                r -= 1

                

            return True
        

        def backTracking(index):
            if index >= len(s):
                res.append(curPart.copy())
                return
            
            for j in range(index, len(s)):
                if checkPali(s[index:j+1]):
                    curPart.append(s[index:j+1])
                    backTracking(j + 1)
                    curPart.pop()

        backTracking(0)
        return res

```
