### 14. Longest Common Prefix

```python
"""
U: we are to return the longest common prefx along all the given element of the array
M: 

- Here one of the restriction is the common prefix among all the given element 
- which means if we choose a single elemnet
- travrse throughth second elmenet and append the common prefix
- after encounter a non-common prefix rethrn the appended string

"""

class Solution(object):
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype str
        """

        select = strs[0]
        res = ""

        for idx in range(len(strs[1])):
            if strs[1][idx] == select[idx]:
                res.append(select[idx])

        return res

```
