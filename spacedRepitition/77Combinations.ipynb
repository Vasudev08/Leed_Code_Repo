```python
"""UMPIRE

U: we are given two numbers 'n' and 'k'
    'n' represents the range [1, n]
    'k' tells us how long the combination has to be

M: Backtracking
NOTE: [1,2] is equal to [2,1] to avoid duplication
we are going to traverse the list from left and right
which way we can avoid duplicates

P:
1. create a backtracking fuction which takes 'startIndex' and 'curComb'
2. if len('curComb') is equal to 'k' then we append it to the 'res' list
3. we will use 'for' loop and pop function to remove and add element
    into the list


"""
class Solution:
    def combine(self, n: int, k: int) -> List[List[int]]:
        res = []

        def backTracking(startIndex, curComb):
            if len(curComb) == k:
                res.append(curComb.copy())
                return
            
            for i in range(startIndex, n + 1): # our range starts from 1
                curComb.append(i)
                backTracking(i + 1, curComb)
                curComb.pop()


        backTracking(1, [])
        return res
```
