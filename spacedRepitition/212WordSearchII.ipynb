### Time Complexity
- TrieNode -> O(n * k), where n: no. of words, k: avg. len of a word
- Outer Loop -> O(m * n), where m: no. of rows, n: no. of cols
- dfs/Backtracking -> O($4^k$) on avg. , k: max length of word and for 
every character in that word we explore 4 possible combinations

- Total : O(n * k + m * n * $4^k$)

### 1st Repetition

```python
"""
We have used TrieNode data structure to help us optiimize the brute force solution
as by constructing the TrieNode data structure we can use it as a filter for our dfs backtracking algo.
if our cur-character doesn't belong in the TrieNode in we can just return or not explore that branch at all
saving us computation time 

TrieNode + dfs + backTracking
"""

class TrieNode:
    def __init__(self):
        self.children = {}
        self.isWord = False

    def addWord(self, word):
        cur = self

        for c in word:
            if c not in cur.children:
                cur.children[c] = TrieNode()

            cur = cur.children[c]

        cur.isWord = True


class Solution:
    def findWords(self, board: List[List[str]], words: List[str]) -> List[str]:
        
        root = TrieNode()
        for w in words:
            root.addWord(w)


        rows = len(board)
        cols = len(board[0])

        res = set()
        visited = set()

        def dfs(r, c, node, word):
            if (r < 0 or c < 0 or r == rows or c == cols or (r, c) in visited or board[r][c] not in node.children):
                return

            visited.add((r, c))

            node = node.children[board[r][c]]
            word += board[r][c]

            if node.isWord:
                res.add(word)

            dfs(r - 1, c, node, word)
            dfs(r + 1, c, node, word)
            dfs(r, c - 1, node, word)
            dfs(r, c + 1, node, word)

            visited.remove((r, c)) # backtracking to explore all possible combination of words


        for r in range(rows):
            for c in range(cols):
                dfs(r, c, root, "")

        return list(res)

            
        
```

```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.isWord = False

    def addWord(self, word):
        # Adding new words into Trie
        cur = self

        for char in word:
            if char not in cur.children:
                cur.children[char] = TrieNode()

            cur = cur.children[char]

        cur.isWord = True

class Solution:
    def findWords(self, board: List[List[str]], words: List[str]) -> List[str]:
        # UMPIRE
        # U: We are given a 2D board with alphabets and a list of words. our aim is to
        # return a list of words which are present in 2D board from the given list of words

        # M: Trie & BackTracking + DFS
        # We use Trie to reduce the searchSpace which optimizes the algorithm
        # how did we reduce the search space ?
        #   - by using we don't have to run the dfs function for every word in words list
        #   - due to data strucutre of trie we can explore mutliple words simulately
        #   - if they have same prefix 

        #P: 
        # - anything word/string will requires Trie Class

        # - Trie Class
        #   - __init__(self)
        #   - addWord function


        # Outside Loop
        #   - backTracking function
        #       - backTracking to explore all possible routes to make a word 
        #           taking trie as a parameter
        # for loop to traverse the given list of words

        rows = len(board)
        cols = len(board[0])
        visited = set()
        res = set()
        
        # Making and adding words to our Trie
        root = TrieNode()
        for w in words:
            root.addWord(w)
            


        def backTracking(r, c, trie, curWord):
            if r < 0 or r >= rows or c < 0 or c >= cols or (r, c) in visited or board[r][c] not in trie.children:
                return

            visited.add((r, c))

            trie = trie.children[board[r][c]]
            curWord += board[r][c] 

            if trie.isWord:
                res.add(curWord)

            backTracking(r + 1, c, trie, curWord)
            backTracking(r, c + 1, trie, curWord)
            backTracking(r - 1, c, trie, curWord)
            backTracking(r, c - 1, trie, curWord)

            visited.remove((r, c))


        for r in range(rows):
            for c in range(cols):
                backTracking(r, c, root, "")

        return list(res)

            



    
        
```
