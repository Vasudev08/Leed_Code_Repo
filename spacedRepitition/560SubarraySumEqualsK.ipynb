```python
"""
UMPIRE:

U: we are given a list of integers and we are to return the count or no. of 
subarrays which equal to 'k'

M: sliding window + for + while loop

P:
- increase the legnth of the window 'r' by the for loop
- decrease the length of the window 'l' when curWindowSum > k 
    and when curWindowSum < k increase the windowLength
- when curWindowSum == k; increment the count of res

"""

class Solution(object):
    def subarraySum(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: int
        """

        resCount = 0
        l = 0
        curWindowSum = 0

        for r in range(len(nums)):
            cur = nums[r]
            curWindowSum += cur

            if curWindowSum == k:
                resCount += 1

            while curWindowSum > k:
                prev = nums[l]
                curWindowSum -= prev
                l += 1

        return resCount
```
