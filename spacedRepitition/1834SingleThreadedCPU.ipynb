```python
# we are adding to heap, when curTime == enqueueTime
# and our minHeap is arranged by the processingTime

# Q1. I need to check whether we can specifcy by which quantity
# the minHeap should work

# here the minHeap is same as the avaible tasks
# here we have to keep track of time t by the use of while loop
# and a count to keep track of processed requested or queued request
# a resList to store the index of the tasks ( we can use minHeap to keep
# track of this as well)

# tasks = [[1,2],[2,4],[3,2],[4,1]]
# While minHeap:
# curTime =1, minHeap = [[processingTime, index], [2, 0]]
    # CPU == IDLE, so pop from minHeap, minHeap = [[processingTime, index]]
    # CPU == BUSY, ProcessingTime = 2, resList = [0]

# curTime = 2, minHeap = [[4, 1]]
    # CPU == BUSY, ProcessingTimeLeft = 1, resList = [0]

# curTime = 3, minHeap = [[2, 2], [4,1]]
    # CPU = IDLE, so pop from minHeap, minHeap = [[4, 1]]
    # CPU = BUSY, processingTimeLeft = 2, resList = [0, 2]

# curTime = 4, minHeap = [[1, 3], [4, 1]]
    # CPU = BUSY, processingTimeLeft = 1

# curTime = 5, minHeap = [[1, 3], [4, 1]]
    # CPU = IDLE, so pop from minHeap, minHeap = [[4, 1]]
    # CPU = BUSY, processingTimeLeft = 1, resList = [0, 2, 1]

# curTime = 6, minHeap = [[4, 1]]
    # CPU = IDLE, so pop from minHeap, minHeap = []
    # CPU = BUSY, processingTimeLeft = 4, resList = [0, 2, 1, 4]

# EXIT WHILE LOOP

# return resList

# time complexity - O(n)
# space complexity - O(n)
import heapq
# if the tasks aren't arranged or sorted, we first sort it

def findTaskOrder(tasks):
    minHeap = []

    curTime = 0
    taskIndex = 0
    cpuAva = False
    processingTimeLeft = 0
    resList = []

    while len(resList) != len(tasks):
        while taskIndex < len(tasks) and curTime == tasks[taskIndex][0]:
            heapq.heappush(minHeap, [tasks[taskIndex][1], taskIndex])
            taskIndex += 1

        if processingTimeLeft == 0:
            cpuAva = True

        if cpuAva and minHeap:
            processingTime, curTaskIdx = heapq.heappop(minHeap)
            processingTimeLeft = processingTime
            resList.append(curTaskIdx)
            cpuAva = False

        curTime += 1
        if processingTimeLeft > 0:
            processingTimeLeft -= 1

        
    return resList

tasklist = [[19,13],[16,9],[21,10],[32,25],[37,4],[49,24],[2,15],[38,41],[37,34],[33,6],[45,4],[18,18],[46,39],[12,24]]
print(findTaskOrder(sorted(tasklist)))
```
