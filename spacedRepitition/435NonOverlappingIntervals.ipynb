### 1st Repetition 2nd Try

```python
"""
UMPIRE:
U: we have been given a list of intervals, and we have to return list of non-overlapping intervals where 
    specifically we want to remove minimum number of intervals to make the list of interval non-overlapping

M: Greedy Approch, where once we find out an overlapping intervals, we remove the one with higher Right Bound

NOTE: an overlapping intervals are when [x, y], [p, q]
        when y > p means overlapping

P:
0. sort the list of intervals takes O(n log n)
1. traverse through the intervals and record the prevEnd of the previous intervals
2. compare the prevEnd with the newIntervals lower bound
3. find out whether it's overlapping or not
    3.1 if overlapping: increment the value of res and also update the prevEnd to the min of both right bounds
    3.2 if not-overlaping: update the prevEnd

"""

class Solution:
    def eraseOverlapIntervals(self, intervals: List[List[int]]) -> int:
        intervals.sort()
        res = 0 # recording the no. of removed intervals

        prevEnd = intervals[0][1]

        for start, end in intervals[1:]:
            if prevEnd <= start:
                prevEnd = end
            else:
                res += 1
                prevEnd = min(prevEnd, end)

        return res

        
```

### 2nd Try - 6/13/2024

```python
"""
UMPIRE:

U: We are given a list of intervals , where we have to remove min.
of intervals to make the list of intervals non-overlapping

M: Greedy Solution, we choose the interval with the lower bound to keep 
    in the list of 'intervals'

overlap when [x, y], [p, q] -> y < q

P:
1. first step is to sort the list of intervals
2. traverse throught the list of intervals and also record
    the prevEnd and compare it with next one
3. when an overlap exists, update the prevEnd accordingly
and increase the no. of removed intervals
"""

class Solution:
    def eraseOverlapIntervals(self, intervals: List[List[int]]) -> int:
        intervals.sort()
        prevEnd = intervals[0][1]
        res = 0

        for l, r in intervals[1:]:
            if prevEnd <= l:
                prevEnd = r
            else:
                res += 1
                prevEnd = min(prevEnd, r)

        return res
    

    # Running time - O(nlog n)
    # constant space complexity
```
