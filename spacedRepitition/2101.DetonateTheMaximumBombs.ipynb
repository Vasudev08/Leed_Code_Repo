### 2101. Detonate the Maximum Bombs - 12/12/2025

```python
class Solution(object):
    def maximumDetonation(self, bombs):
        """
        :type bombs: List[List[int]]
        :rtype: int

        So, I tried solving it but I also found a better way to solve it as well
        where we at the beginning of the problem we construct a 
        adjacenyList, which contains the idx of a each bomb and which other
        bomb can be reached from the current idx bomb

        Ran running a dfs function to get the Detonate the maximum bombs
        """

        adjList = defaultdict(list)
        # this tells which possible circles we can look at
        n  = len(bombs)

        for i in range(n):
            for j in range(n):
                if i == j: continue
                x1, y1, r1 = bombs[i]
                x2, y2, r2 = bombs[j]

                if r1**2 >= (x1 - x2)**2 + (y1 - y2)**2:
                    adjList[i] += [j]

        print(adjList)

        def dfs(node, visited):
            for ch in adjList[node]:
                if ch not in visited:
                    visited.add(ch)
                    dfs(ch, visited)


        ans = 0
        for i in range(n):
            visited = set([i])
            dfs(i, visited)
            ans = max(ans, len(visited))

        return ans



        
        
```
