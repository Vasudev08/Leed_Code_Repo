```python
"""
UMPIRE

U: we are given n, which tells us how many pair of parentheses we have
our job is to return a list of strings of well-formed parentheses

well-formed parentheses list:
"()", "()()"

Non well-formed parentheses list:
")()"

M: backtracking to explore all possible combination, and as we 
need a string, we will only use single list to keep track and use backtracking

1. Here, our first element can't be an ")" close parentheses, as we 
wouldn't be able to completely close it afterwards
2. to keep track of use of close and open parentheses, we will have two 
variables called "openN", "closeN"
3. when our "openN" is equal to "closeN" equal to "n", then we have a valid
well-formed parentheses
4. we can add open parentheses, when openN is less than n
5. and we can only add close parentheses, when closeN < openN, this is because
it becomes impossible to make it an complete parentheses if we start with an
")" closing parentheses

"""

class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        res = []
        curString = []

        def backtracking(openN, closeN):
            if openN == closeN == n:
                res.append( "".join(curString))


            if openN < n:
                curString.append("(")
                backtracking(openN + 1, closeN)
                curString.pop()

            if closeN < openN:
                curString.append(")")
                backtracking(openN, closeN + 1)
                curString.pop()


        backtracking(0, 0)
        return res

```

```python
class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        res = []
        par = [')', '(']

        def backTracking(curStr, curOpen):
            if len(curStr) == 2 * n and curOpen == 0 and curStr[-1] == ')':
                res.append(curStr)

            if len(curStr) > 2 * n:
                return

            for i in par:
                if i == ')' and curOpen == 0:
                    continue
                curStr += i
                if i == '(':
                    tempOpen = curOpen + 1
                else:
                    tempOpen = curOpen - 1
                backTracking(curStr, tempOpen)
                curStr = curStr[:-1]
        

        backTracking("(", 1)
        return res
        
```
