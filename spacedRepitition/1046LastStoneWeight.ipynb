```python
# You are given an array of integers stones where stones[i] is the weight of the ith stone.

# We are playing a game with the stones. On each turn, we choose the heaviest two stones and smash them together. Suppose the heaviest two stones have weights x and y with x <= y. The result of this smash is:

#     If x == y, both stones are destroyed, and
#     If x != y, the stone of weight x is destroyed, and the stone of weight y has new weight y - x.

# At the end of the game, there is at most one stone left.

# Return the weight of the last remaining stone. If there are no stones left, return 0.

 

# Example 1:

# Input: stones = [2,7,4,1,8,1]
# Output: 1

"""
Here we are going to use the priority heap for sure, or minHeap and we are going to store the values in negative value
and pop it twice to and follow the aboev rules and put it back on the min heap as well

"""

import heapq


def stoneWeight(stones):
    pq = []

    for w in stones:
        heapq.heappush(pq, -w)

    
    while len(pq) > 1:
        fw = -1 * heapq.heappop(pq)
        sw = -1 * heapq.heappop(pq)

        if fw != sw:
            nw = abs(fw - sw)
            heapq.heappush(pq, -nw)

    return -1 * heapq.heappop(pq) if len(pq) > 0 else 0
    


stoneWeight([2,7,4,1,8,1])
```
