```python
"""
this problem is similar to climbing stairs and robbing house 
problem in a way, 
- where in the climibing stairs problem 
    we are climbing 1 or 2 step as a time

- in house robbing problem
    we can rob the 2nd house or 3rd or 4 rd from the current house

- in this problem we can choose an digit as individual
    OR as two digit


So, 

1   2   3
Here, last element 3 can be decoded in 1 way: 3
second last element 2 can be decoded in 2 way : 23, 2
first elment 1 can be decoded in 2 way: 1, 12
    and if we choose 1, we have to find way to decode 23
    which are 2 so, in total we have 3 total way to decode the string

"""

class Solution:
    def numDecodings(self, s: str) -> int:

        dp = {len(s): 1}

        def dfs(i):
            if i in dp:
                return dp[i]
            
            if s[i] == '0':
                return 0
            
            res = dp(i + 1)
            if (i + 1 < len(s) and (s[i] == '1' or s[i] == '2' and s[i + 1] in '0123456')):
                res += dfs(i + 2)

            dp[i] = res

            return res
        
        return dfs(0)
```
