```python
"""
2nd Try: this implementation is little bit difficult to undersand we 
it's difficult to visualize the implementation so I recommend first implementating 
a different approach where we are using a hashmap (or you can say memory)

The difference between house robber and climbing stairs problem in the 
dyanmic programming folder is that in climbing stairs problem we restrict
ourself to 1 step or 2 step jump

whereas in the house robber, we can choose the 2+ step houses 
which makes alll the difference
"""

class Solution:
    def rob(self, nums: List[int]) -> int:
        rob1, rob2 = 0, 0

        for n in nums:
            temp = max(n + rob1, rob2)
            rob1 = rob2
            rob2 = temp

        return rob2
    

class Solution2:
    def rob(self, nums: List[int]) -> int:
        dp = [0 for i in len(nums) + 1]

        dp[0] = 0 # no house on money can be robbed
        dp[1] = nums[0] # the maximum amount of money that can be stolen from
                        # first house is just nums[0]
        
        for i in range(2, len(nums) + 1):
            dp[i] = max(dp[i - 1], dp[i - 2] + nums[i - 1]) # nums is i - 1 because of zero index

        return dp[-1]
```

```python
"""
UMPIRE:
U: given a list of houses having certain amount of money, 
    return the maximum money that can be stolen 
    without robbing two adjacent houses at the night

M: Recursion, hashmap
similar sa the total no. of ways of climbing stairs
    AND min cost of climibing stairs

LOGIC: starting from the end of the list, calculate the that can be stolen
at the current place 
    - which means current place money + money that can be stolen at 
        place current place + 2 
    - so if we are traversing from the end of the list
    - last house/element = curMoney
    - 2nd last house/element = curMoney
    - 3rd last house/element = curMoney + last house money
    and so on

P:
1. create a helper function, which takes the current house no
    and returns the money that would be stolen at the current place

    i j 
    3 6
    NOTE:
    - the code fails for the following cases:
        [2,1,1,2]
        0, 0, [1, 3, 1, 3, 100]
        when i = 100, I can either choose max till the element 3 or max of [1, 3, 1]
        and we store max in rob2

        where we have to choose any houses more than currentHouse + 2


"""

class Solution:
    def rob(self, nums: List[int]) -> int:
        moneyMap = {}

        def maxMoney(i):
            if i >= len(nums):
                return 0
            
            if i in moneyMap:
                return moneyMap[i]
            
            moneyMap[i] = nums[i]

            tempMoney = nums[i] + maxMoney(i + 2)

            return tempMoney
        
        return max(maxMoney(0), maxMoney(1))
            
```
