### 2nd Try

```python
"""
1st Try Implementation gives an exceed runtime error so we have to use memoization 
to use memoization in this problem we have to understand that 
- To reach the top from the last staircase - there is only one way to do it
- To reach the top from the 2nd last staircase - there is either jumping to top (first way)
    or going through the last staircase/step.
    which means total unqiue ways to reach tops from 2nd last staircase is 
    1 + waystoreachtop from last staircase/step = 1 + 1 = 2
- and the story cotinues and please look at the above diagram
"""

class Solution:
    def climbStairs(self, n: int) -> int:
        memo = {}        

        def uniqueWay(curStep):
            if curStep == n:
                return 1

            if curStep > n:
                return 0

            if curStep in memo:
                return memo[curStep]

            ways = uniqueWay(curStep + 1) + uniqueWay(curStep + 2)
            memo[curStep] = ways
            return ways            
       

        return uniqueWay(0)
        
```

### 1st Try

```python
"""
UMPIRE
U: each time we can either climb 1 or 2 steps.
    given n, where it takes n steps to reach the top.
    return total no. of distinct way to reach to the top

    stairs:
    1  2  3  4  5  6  7
    *
       * 
           *
       -

At the beginning we have two choices
either to start at 1st Step
or to start at 2nd Step

Afterwhich on every step we have two choices

M: Recursion + dfs
NOTE we are starting with one, '1'
P:

- build a recursion function
    - BASE CASE I: when curStep is equal to n, then increase the wayCount by 1
    - BASE CASE II: when curStep is more than n, then just return
    - Recursion Call I: call the recursion function taking +1 step
    - Recursion Call II: call the recursion function taking +2 step

- Call the recursion funtion starting from 1st step
- Call the recursion function starting from 2nd step

O(n) -> running time
O(1) -> space complexity


"""

class Solution:
    def climbStairs(self, n: int) -> int:
        wayCount = 0
        def unquieWay(curStep):
            if curStep == n:
                wayCount += 1

            if curStep > n:
                return
            
            unquieWay(curStep + 1)
            unquieWay(curStep + 2)

        unquieWay(1)
        unquieWay(2)
                
```
