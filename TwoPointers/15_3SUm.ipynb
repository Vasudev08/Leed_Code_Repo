```python
"""
3Sum is the evoluation of 2Sum, and here we are going to convert 3sum to 
an 2sum problem by using while and for loop

U: given an array, return an triplet summing up to zero
M: Two pointers + Hashing (Hashing to check if an number exists in O(1) running time)
P:
1. have two pointers named - pri and sec
2. pri points to first number in the triplet, this will be accmplished by using
while loop to make sure it's doesn't cross the length of the array
3. after entering the while loop, we can use sec pointer to point to the second 
element int he triplet and check in the potential thrid element exists in the hashmap
4. if it exists, then we append the three values into a res list and add them to visited
so that we don't have duplicate triplets.


"""

class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        pri = 0
        target = 0
        num_map = {}
        res = 0

        for i, n in nums:
            num_map[n] = i


        while (pri < len(nums)):
            visited = set()
            visited.add(pri)
            cur_target = target + nums[pri]
            sec = pri + 1
            while (sec < len(nums)):
                thi = cur_target + nums[sec]
                if (thi in num_map and thi not in visited):
                    # To - Do update 
                    visited.add(thi)
                    visited.add(sec)
                    
                    res.append([nums[pri], nums[sec], nums[thi]])

                sec += 1

        return res
            
```
