### 128. Longest Consecutive - 2/6/2024

```python
"""
UMPIRE:
U: given an unsorted array, return the largest consective sequence length
    - Here we can't use sorting alogrithm because the running time of sorting is 
    - O(nlogn)

M: Using hashing particularly dictionary and visted_set()
P:
- store the occurence of each element in a res hashmap
- traverse through the unsorted array
    - for current element, check if it's incrementment exists in res and not in visited_set()
    - if so, go into the while loop where we will get the length 
    - of sequence and add to the visited_set()
    - compare the local length with max_length and store the max_length
- return length

"""
class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        visited = set()
        res = {}
        max_len = 0

        # Recording the occurence of every element
        for n in nums:
            res[n] = 1


        for i in range(len(nums)):
            if (nums[i] + 1 in res) and (nums[i] not in visited):
                visited.add(nums[i])
                loc_len = 1
                cur = nums[i] + 1
                
                while (cur in res):
                    visited.add(cur)
                    loc_len += 1
                    cur += 1
                
                max_len = max(max_len, loc_len)
            else:
                max_len = max(max_len, 1)

        return max_len
                    


# The above algorithm is slow and inefficient
```

```python
"""
Improvements:
- we can use set() rather than dictionary
    - we used dictionary for checking if num + 1 existed or not
    - but that can be done using if num - 1 not in set() --> o(1) operation
    - set have overall faster time

- by checking (nums - 1 not in nums_set) be eleminated any need for
- requiring an visited set at all
- i.e. array = [3, 4, 5, 9, 10]
when num = 4, then the loop will stop because num -1 = 3 exits in the list or set


"""

class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:

        if not nums:
            return 0
        
        nums_set = set(nums)
        max_length = 0

        for n in nums_set:
            if n - 1 not in nums_set:
                cur_length = 1
                cur = n

                while cur + 1 in nums_set:
                    cur_length += 1
                    cur += 1

                max_length = max(max_length, cur_length)

        return max_length
            
```
