```python
"""
UMPIRE:
U: given a list of courses where [i, j] for to take i class the pre-req is j
and we have to return true if all the coures can be taken (i.e. there is 
no cycle in the pre-rep and courese to be taken)
for example [[1, 0], [0, 1]]: to take course 1 we need to complete course 0. AND
to take course 0 we need to complete coure 1 which is a cycle and hence the return
statement is false
M: dfs + hashMap + detect cycle using visitedSet
P:
- create a hashmap of crs and their pre-req
    - courses havnig no pre-req, will have empty list as their value
        which means they can be taken 
        and any courses have this course as a pre-req can also be taken
    - courses have pre-req will have a list of pre-req courses on which
        we will call the dfs function

    [] : represents can be taken without any problem
    if current crs in visitedSet() means we detected a cycle  
    hence return False

- if we find out that a course can be completed even after
    having pre-req we will replace its value to empty list [] to indicate that
    and make the algo run faster


- define a recursion function having current course as a parameter
    - Base Case I: if crs in the visitedSet: return False
    - Base Case II: if hashmap[crs] == []: return True

    add the current course to the hashmap
    call the recursion function on all of it's pre-req courses
    after processing remove the current course from the visitedSet to avoid errors
    (to false detection of cycle) and replace the hashmap current crs value to 
    empty list

    return True


    Do a visual dry run to understand the implementation more please

"""

class Solution:
    def canFinish(self, numCourses: int, prerequisites: List[List[int]]):
        preMap = {i : [] for i in range(numCourses)}
        visitSet = set()

        for crs, pre in prerequisites:
            preMap[crs].append(pre)

        def dfs(crs):
            if crs in visitSet:
                return False

            if preMap[crs] == []:
                return True

            visitSet.add(crs)
            for pre in preMap[crs]:
                if not dfs(pre):
                    return False

            visitSet.remove(crs)
            preMap[crs] = []
            return True

        for crs in range(numCourses):
            if not dfs(crs):
                return False

        return True



```
