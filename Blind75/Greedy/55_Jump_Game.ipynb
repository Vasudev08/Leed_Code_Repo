```python
"""
UMPIRE:

U: given a nums array, ith element resprsent possible jumps we can make,
return true if we can reach the last index
Q1. let's us say nums[i] = 3, is that telling us we can jumps once or twice or thrice 
    or only thrice ?
A1. element values represnt the maximum jump value
NOTE: Our aim is to reach the last position

M: Dynamic Programming + traversing from the end of the array

P: See Picture
- create a dp hashmap having values 0
    which represents false, which means we can't reach the end of the list
    from that position either through the element or directly
- traverse from the end of the array while checking if we can reach the end of the list
    and update the hashmap
- return dp[0] if 0 return true else false
"""

class Solution:
    def canJump(self, nums: List[int]) -> bool:
        dp = [0] * len(nums)
        dp[-1] = 1

        for i in range(len(nums) - 2, -1, -1):
            tmp = nums[i]
            if (i + 1 + tmp) >= len(nums) or (1 in dp[i: (i + tmp + 1)]):
                dp[i] = 1

        return True if dp[0] == 1 else False
```
