```python
# TRY 1

"""
UMPIRE

U: return true if subroot is in root, return false if subroot isn't inside the root
M: dfs (preOrder) + bfs (Que + While)
P:

- create a helper function preOrder
    - find the first match between the root and subroot
    - Once match found
        - implement an bfs to check if all the node are the same
        - if so, return True
        - else return False
"""

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def isSubtree(self, root: Optional[TreeNode], subRoot: Optional[TreeNode]) -> bool:
        # bfs + dfs
        # bfs -> find the first match
        # dfs -> using que and while loop see if both trees are same
            # if so, return True
            # else return False

        def preOrder(node, subNode):
            if not node:
                return

            if node.val == subNode.val:
                que = [node]
                que2 = [subNode]
                while que:
                    queLen = len(que)
                    for i in range(queLen):
                        tmp = que.pop(0)
                        tmp2 = que2.pop(0)
                        if tmp.left and tmp2.left and tmp.left.val == tmp2.left.val:
                            que.append(tmp.left)
                            que2.append(tmp2.left)
                        else:
                            return False

                        if tmp.right and tmp2.right and tmp.right.val == tmp2.right.val:
                            que.append(tmp.right)
                            que2.append(tmp2.right)
                        else:
                            return False

                return True

            return (preOrder(node.left, subNode) or (preOrder(node.right, subNode)))


        return (preOrder(root, subRoot))
        
```
