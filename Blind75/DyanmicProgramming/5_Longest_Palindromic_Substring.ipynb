```python
"""
2nd Try
 Q1. how we are going to detect a palndrom
    - Using two pointers
        - if a palindrome is given we can check if it's so
        by having one pointer starting at the beginning
        another one starting at the end. and after checking
        these two pointers value we are going to increase left pointer
        and decrease right pointer
    - In a odd-length palindrome - middle value represents the center
        and we can keep on expanding left and right until left != rigth
    - In a even-legnth palindrome - two middle value represents the center
        and we can keep on expanding left and right until left != right
    NOTE:
        - There are two way to check for paldindrome
        1. from the middle
        2. from the ends

"""

class Solution:
    def longestPalindrome(self, s: str):
        res = ""
        # ODD Length
        for i in range(len(s)):
            l, r = i, i

            while (s[l] == s[r] and l > -1 and r < len(s)):
                l -= 1
                r += 1

            tmp = s[l:r]
            if len(tmp) >= len(res):
                res = tmp

        # EVEN Length
        for j in range(len(s)):
            l, r = i, j + 1

            while (s[l] == s[r] and l > -1 and r < len(s)):
                l -= 1
                r += 1
            
            tmp = s[l:r]
            if len(tmp) > len(res):
                res = tmp

        return res

# You can do this
    
class Solution2:
    def longestPalindrome(self, s: str):
        res = ""

        for i in range(len(s)):
            l, r = i, i

            while (l >= 0 and r < len(s) and s[l] == s[r]):
                if (r - l + 1) > len(s):
                    res = s[l: r + 1]

                l -= 1
                r += 1

            l, r = i, i + 1
            while (l >= 0 and r < len(s) and s[l] == s[r]):
                if (r - l + 1 ) > len(s):
                    res = s[l: r + 1]

                l -= 1
                r += 1

        return res

        

        


```
