```python
"""
NOTE BRUTE FORCE Approach
UMPIRE
U: given a array of numbers, return the product or largest product
    NOTE Here the product has to be from the sub-array
    ex array = [2, 3, -2, 4] subarray = [2, 3] product = 6

    we have to the potential algo on every element and keep track of maxProduct
    Brute Force:
        - use two 'for' loops to traverse through the array and compute the product 
        and record it 

        for i in range(len(nums)):
            l = nums[i]
            for j in range(i + 1, len(nums)):
                pro = l + nums[j]

                res = max(pro, res)

        Please see the below implementation

            

"""

####### BRUTE FORCE APPROACH ##################################
class Solution:
    def maxProduct(self, nums: List[int]) -> int:
        if len(nums) == 1:
            return nums[0]
        res = 0

        for i in range(len(nums)):
            l = nums[i]
            pro = l
            res = max(pro, res)

            for j in range(i + 1, len(nums)):
                pro *= nums[j]
                res = max(pro, res)


        return res


        
```
