```python
"""
        2nd TRY

the logic is correct, the implementation is almost correct
as we made two improvements
1. when i >= len(cost) we are returning 0 as we're already at the top
2. while calling recursion function we are also adding the curstep cost
        by using cost[i] in the recursion call
3. while return in the function we should return the minimum of 
        1st step or 2nd step
"""

class Solution:
    def minCostClimbingStairs(self, cost: List[int]) -> int:
        memo = {}

        def minCost(i):
            if i >= len(cost):
                return 0
            
            
            if i in memo:
                return memo[i]
           
            curCost = cost[i] + min(minCost(i + 1), minCost(i + 2))
            memo[i] = curCost
            

            return curCost
        
        return min(minCost(0), minCost(1))
        
```

```python
"""
        1st TRY
UMPIRE

U: given a list of cost to climb each staircase, return the minimum 
cost to reach the top of the staircase

this question is very similar to 70. leet code question
but the required output for this question is returning
the minimum cost of climbing the stairs

logic behind climbing the staircase in Q70:
from current step calculate way to get to the top (i.e. we are starting from behind)
- so, from the last staircase it's just 1 step
- from the second staircaise it's just either jump 2 steps (1st way)
or go through the jump last step (2nd way) so in total 2 ways

Similary logic behind Min Cost Climbing Stairs 
- from current step calcuate the min cost required to reach the top of staircase

M: recursion
P: 
- start from the ending of the list and calculate at each ith
    the min cost to get to top from that ith step
- So, from the last step it will be 0


"""
class Solution:
    def minCostClimbingStairs(self, cost: List[int]) -> int:
        memo = {}

        def minCost(i):
            if i > len(cost):
                return float('inf')
            
            if i == len(cost):
                return cost[i]
            
            if i in memo:
                return memo[i]
           
            curCost = min(minCost(i + 1), minCost(i + 2))
            memo[i] = curCost
            

            return curCost
        
        return minCost(0)
```
