```python
"""
UMPIRE:
U: given M x N grid where a robot is placed on grid[0][0] and wants to reach grid[m-1][n-1]
    return no. of unique ways the robot can reach the end point (i.e. grid[m-1][n-1])

M: 
Idea: As we can only move to right and down, we can count the no. of ways to get to a tile
from above picture we can say two things
1. No. of ways to get to a cell is adding no. of ways to getting at its left neighbor + top neighbor
2. First row and first column are going to be '1' as there is only 1 way to get their

P.
- create a 2d hashmap known as dp
- make the first row and first column as having value '1'
- traverse through the rest of the grid and calculate currentCell = leftCell + rightCell 
- return dp[m - 1][n -1]

"""

class Solution:
    def uniquePaths(self, m: int, n: int) -> int:
        # m * n m: rows, n: columns
        dp = [[0] *n] *m

        # Only 1 Way to reach first column
        for i in range(m):
            dp[i][0] = 1

        # Only 1 Way to reach the first row
        for j in range(n):
            dp[0][j] = 1

        for i in range(1, m):
            for j in range(1, n):
                dp[i][j] = dp[i - 1][j] + dp[i][j - 1]

        return dp[m-1][n-1]

```
