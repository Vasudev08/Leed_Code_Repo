### 417. Pacific Atlantic Water Flow - 3/2/2024

```python
"""
UMPIRE:
U: we an 2D array of heights, return a list of 2d co-ordinate cell
    from which the water can flow to both oceans
M: dfs
NOTE I know we should start from the ocean
P:
    1. get all the cells from which we can go to pacific ocean
    2. get all the cells from which we can go to altanic ocean
    3. append common cell (i.e. belong to both pacific ocean and altanic ocean)
      into a res list

More Detail:
- rain follows the path of decreasing height, which means
    for a water from a cell to reach both ocean, the water 
    has to follow a path of decreasing height
- we can use the above fact to our advantage and start 
    from the ocean and make note of cells from which we 
    can reach that particular ocean 
- which can be done by taking note of cells starting from ocean
    with increasing height. for example cell with value 4 is near to the ocean
    then the next potential cell which can reach the ocean is cells with values
    4 or more. and we make note of these cells
- we make note of cells for both side of the ocean (i.e. pacific ocean and altanic ocean)
    and ouput the comon cells among them
- In our 2D-array top and left side are pacific ocean
    which means all the top row and leftest column cells can reach paccific ocean
    and from each cells we need to make note of all possible cells with equal or increaing
    heights.
- Similarly, 2D-array bottom and righest side are near to atlantic ocean
    which mean all the rightest and bottom cells can reach the atlantic ocean
    and from each of these nodes we need to make note of cells with equal or increasing 
    height
"""

class Solution:
    def pacificAtlantic(self, heights: List[List[int]]) -> List[List[int]]:
        pac, atl = set(), set()
        rows = len(heights)
        cols = len(heights[0])

        def dfs(r, c, visit, prevHeight):
            if r < 0 or c < 0 or r >= rows or c >= cols or (r, c) in visit or heights[r][c] < prevHeight:
                return 
            
            visit.add((r, c))

            dfs(r + 1, c, visit, heights[r][c])
            dfs(r - 1, c, visit, heights[r][c])
            dfs(r, c + 1, visit, heights[r][c])
            dfs(r, c - 1, visit, heights[r][c])


        for c in range(cols):
            dfs(0, c, pac, heights[0][c])
            dfs(rows - 1, c, atl, heights[rows - 1][c])
        

        for r in range(rows):
            dfs(r, 0, pac, heights[r][0])
            dfs(r, cols - 1, atl, heights[r][cols - 1])


        res = []
        for r in range(rows):
            for c in range(cols):
                if (r, c) in pac and (r, c) in atl:
                    res.append([r, c])


        return res
            

            
```
