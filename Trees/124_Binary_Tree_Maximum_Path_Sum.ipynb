### 124. Binary Tree Maximum Path Sum - 2/20/2024

```python
"""
2nd Try

we are taking an different approach towards solving this problem.
we are using to use recursion 
BASIC IDEA: we need to calculate the Max of left subtree and right subtree
without the split.
We will also have an res global variable where we store the maximum path sum
and while updating/ assignign it, we can the current Node and split our path 
here. i.e. (we can only have an occurence of 1 per path). that's why we can
only affoard 1 split. which has to be at the top.)
"""

class Solution:
    def maxPathSum(self, root: TreeNode) -> int:
        res = [root.val]

        def dfs(root):
            if not root:
                return 0

            leftMax = dfs(root.left)
            rightMax = dfs(root.right)
            leftMax = max(leftMax, 0) # when we have negative numbers
            # we don't have to choose them 
            rightMax = max(rightMax, 0)

            res[0] = max(res[0], root.val + leftMax + rightMax)

            return root.val + max(leftMax, rightMax)

        dfs(root)
        return res[0]
```

```python
"""
There's an better approach to accomplish this for sure




UMPIRE
U: given a binary tree, return the maximum path sum, 
NOTE can only use the node once

M: Hashmap to up the tree, or go to parent node
    visitedSet as well

P: 
- have to start the pathSum Algo from each node
- where each node has three choices at every step
    1. go to parentNode -> (Using parentMap)
    2. go to leftNode
    3. go to rightNode
"""

class Solution:
    def maxPathSum(self, root: Optional[TreeNode]) -> int:



        def getParentMap(node, parent):

            parentMap[node] = parent
            if node.left:
                getParentMap(node.left, node)
            if node.right:
                getParentMap(node.right, node)


        

        def helper(node, pathSum, setVisited):
            if not node:
                return 
            
            maxSum = max(maxSum, pathSum)

            if parentMap[node] not in setVisited:
                setVisited.add(node.val)
                helper(parentMap[node], pathSum + node.val, setVisited)

            if node.left and node.left not in setVisited:
                setVisited.add(node.val)
                helper(node.left, pathSum + node.val, setVisited)

            if node.right and node.right not in setVisited:
                setVisited.add(node.val)
                helper(node.right, pathSum + node.val, setVisited)

            if node.left:
                helper(node.left, node.val, (node.val))

            if node.right:
                helper(node.right, node.val, (node.val))


        maxSum = 0

        parentMap = {}
        getParentMap(root, None)
        helper(root, root.val, (root.val))

        return maxSum

        
```
